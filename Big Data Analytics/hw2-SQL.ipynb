**TODO**: Using `exploded_movies_df` and `streaming_df`:
- Find all movies on Netflix, then **for each genre,** identify the ones with the highest combined RT tomatometer and audience rating (include all ties). Call this new column `combined_rt_rating`.
- Rename the key `genre_category`, and sort your final result by this column in ascending order.

Though it could be helpful, **you do NOT have to implement this in pandas**. Store the results in `netflix_best_df`, which should have the following format:

>movie_title | genres | genre_category | combined_rt_rating
>--- | --- | --- | ---

where `genres` represents all the genres of the movie, and `genre_category` represents the genre that the movie's RT ratings are compared to.

```python
exploded_movies_df.head(3)
```

```python
# TODO: use pandasql to find the best on Netflix
netflix_best_query = """SELECT movie_title, genres, genre AS genre_category, (tomatometer_rating + audience_rating) AS combined_rt_rating
                        FROM  exploded_movies_df
                        WHERE movie_title IN (SELECT Title FROM streaming_df WHERE Netflix = 1)
                        GROUP BY genre_category
                        HAVING combined_rt_rating >= MAX(combined_rt_rating)
                        ORDER BY genre_category ASC
"""
netflix_best_df = ps.sqldf(netflix_best_query, locals())
```

```python
netflix_best_df
```

```python
# TODO: use pandasql to find the best on Netflix
netflix_best_query = """SELECT t1.movie_title, t1.genres, t1.genre AS genre_category, (t1.tomatometer_rating + t1.audience_rating) AS combined_rt_rating
                        FROM  (SELECT genre, max(tomatometer_rating + audience_rating) AS rt 
                               FROM exploded_movies_df 
                               WHERE movie_title IN (SELECT Title FROM streaming_df WHERE Netflix = 1) 
                               GROUP BY genre) t2
                        JOIN exploded_movies_df t1
                        ON (t1.tomatometer_rating + t1.audience_rating) = t2.rt AND t1.genre = t2.genre
                        WHERE movie_title IN (SELECT Title FROM streaming_df WHERE Netflix = 1) 
                        ORDER BY genre_category ASC
"""
netflix_best_df = ps.sqldf(netflix_best_query, locals())
```

Although all critics on RT are officially approved, we want to establish a top-critic-only score that only includes the opinions of the "cream of the crop". Certain films could be poorly received by regular critics, but top critics see something special in them. Let's try and find these underappreciated movies!

**TODO**: Using `reviews_df` and `movies_df`:

- Constrain the movie pool to **only ones that have more than 5 top critic reviews** to mitigate outlier impact.

>movie_title | top_critic_tomatometer | tomatometer_rating | score_diff
>--- | --- | --- | ---

**Hints:**

```python
# TODO: use pandasql to find the regular critics' underrated movies
underrated_movies_query = """
                          SELECT movie_title, 
                                 top_critic_tomatometer,
                                 tomatometer_rating,
                                 (top_critic_tomatometer-tomatometer_rating) AS score_diff
                          FROM (SELECT count(*) AS critic_n, 
                                       movie_title, 
                                       (sum(CASE review_type WHEN 'Fresh' THEN 1 ELSE 0 END)*100/count(review_type)) AS top_critic_tomatometer,
                                       tomatometer_rating
                                 FROM (SELECT *
                                       FROM reviews_df t1 
                                       JOIN movies_df t2 
                                       ON t1.rotten_tomatoes_link = t2.rotten_tomatoes_link
                                       WHERE t1.top_critic = 1)
                                  GROUP BY rotten_tomatoes_link)
                          WHERE critic_n > 5
                          ORDER BY score_diff DESC
                          LIMIT 10  
"""
```

```python
import nltk
nltk.__version__
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
stopwords = set(stopwords.words('english'))
# Note that stopwords are all in lowercase format
```

```python
# TODO: Create tokenized_content(content) function
from nltk.tokenize import word_tokenize
def tokenized_content(content):  
    word_tokens = word_tokenize(content)
    word_tokens = [w.lower() for w in word_tokens if w.isalpha() and not w.lower() in stopwords]
    return word_tokens
```
